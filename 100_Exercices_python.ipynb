```python
## Exercice 37

def calculSomme(L):
    ## initialisation d'une variable somme à 0
    somme = 0
    ## pour chaque nombre dans la liste L
    for nbr in L:
        ## le rajouter à notre variable somme
        somme += nbr
    ## retourner la somme des nombres
    ## obtenus à la fin de la boucle
    return somme

## ---- Appel de Fonctions ---- ##
# calculSomme([3,2,6,9,-1,5])
calculSomme([-3,-6,0,1,2,7])
```

```python
## Exercice 38

def supprimerDoublons(L):
    ## parcourir tous les éléments de la liste L
    for elt in L:
        ## nombre d'occurrences de cet élément 
        ## dans la liste L
        elt_occ = L.count(elt)
        ## S'il existe plus de deux occurrences
        ## de cet élément dans la liste L
        if elt_occ >= 2:
            ## supprimer toutes les occurrences
            ## et garder qu'un seul élément à la fin
            for i in range(elt_occ-1):
                 L.remove(elt)
    ## trier la liste dans un ordre
    ## croissant
    L.sort()
    return L

## ---- Appel de Fonctions ---- ##
# supprimerDoublons([0,3,5,7,3,5,1,-1])
supprimerDoublons([0,5,9,10,3.2,1,-3])
```

```python
## Exercice 40

def maximum(L):
    ## initialiser la variable max_L
    ## avec le premier élément de la liste L
    max_L = L[0]
    ## parcourir tous les éléments de la liste L
    for elt in L:
        ## Si l'élément est plus grand que la
        ## valeur stockée dans max_L
        if elt > max_L:
            ## mettre à jour la variable max_L
            ## avec cet élément
            max_L = elt
    ## retourner la valeur maximale trouvée
    return max_L

### --- Appel de Fonctions ---###
# maximum([-9,2,4,1,8])
maximum([-3,1,7,6,2,3])
```

```python
## Exercice 46

def diviseur(n):
    ## la liste qui va contenir les diviseurs
    n_diviseur = []
    ## parcourir tous les éléments de 1 à n
    for div in range(1,n+1):
        ## si le reste de la division de n par div est 0
        if n%div == 0:
            ## alors div est un diviseur de n
            ## et on le rajoute dans la liste des diviseurs
            n_diviseur.append(div)
    return n_diviseur

## -- Appel de Fonction -- ##
# diviseur(3)
diviseur(9)
```

```python
## Exercice 49

def nbrValeurDict(d):
    ## extraire la liste des clés
    ## dans le dictionnaire d
    d_cle = list(d.keys())
    ## initialisation de la variable
    ## qui va contenir le nombre de valeur
    ## du dictionnaire d
    nbr_valeur = 0
    ## parcourir toutes les clés
    for cle in d_cle:
        ## pour chaque clé, déterminer le nombre
        ## de valeurs contenus dans la liste associée
        ## à cette clé
        longueur_val = len(d[cle])
        ## ajouter ce nombre de valeur à la variable nbr_valeur
        nbr_valeur += longueur_val
    
    return nbr_valeur

## -- Appel de Fonction -- ##
# nbrValeurDict({"a" : [1,2,3], "b" : [3,"p"], "c" : [8]})
nbrValeurDict({"Julie" : [12,60.1], "Fred" : [26,75.6], "David" : []})
```

```python
## Exercice 51

def calculFactorielle(n):
    ## factoriel de n = n x (n-1) x (n-2) x ....x 2 x 1
    
    ## la factorielle du nombre 0 est égale à 1
    if n == 0:
        return 1
    ## la variable qui va contenir
    ## notre résultat final
    res_factoriel = n
    ## on parcourt tous les éléments de n-1 à 1
    for k in range(n-1,0,-1):
        res_factoriel = res_factoriel*k
    
    return res_factoriel
 
## ---- Appel de Fonction ---- ##
# calculFactorielle(0)
# calculFactorielle(9)
calculFactorielle(3)
```

```python
## Exercice 52

def diviseursMult(n,a,nbrSeuil):
    resultat = []
    ## parcourir les nombres de 0 et nbrSeuil
    for nbr in range(nbrSeuil+1):
        ## le reste de la division de "nbr" par "n" doit être égale à 0
        ## et le reste de la division de "nbr" par "a" doit être différente de 0 
        if nbr%n == 0 and nbr%a != 0:
            ## si les conditions sont satisfaites, alors
            resultat.append(nbr)
    return resultat

## --- Appel de Fonction --- ##
# diviseursMult(5,2,100)
diviseursMult(11,3,85)    
```

```python
## Exercice 56

def filtrerMots(phrase,longueurMini):
    ## convertir la phrase en liste en utilisant comme
    ## séparateur un "espace"
    phrase_liste = phrase.split(" ")
    ## la variable qui va contenir notre phrase filtrée
    phrase_filtre = []
    ## parcourir chaque élément de la liste
    for mot in phrase_liste:
        ## Si sa longueur est supérieure à la longueur
        ## minimale passée en paramètre
        if len(mot) >= longueurMini:
            ## alors, on garde ce mot et on l'ajoute dans
            ## notre nouvelle liste
            phrase_filtre.append(mot)
    
    ## convertir la liste en chaine de caractères
    ## en gardant les espaces entre les mots
    phrase_filtre = " ".join(phrase_filtre)
    
    return phrase_filtre

## -- Appel de Fonction -- ##
# filtrerMots("Salut toi",4)
filtrerMots("Quel est ton origine ?",5)
```

```python
## Exercice 58

def nombreOccurrence(L):
    ## Initialisation de la liste dans laquelle 
    ## nous allons stocker notre résultat
    res_tuples = []
    ## parcourir chaque élément de L
    for elt in L:
        ## nombre d'occurrences de l'élément
        nbrOcc = L.count(elt)
        ## création du tuple (élément,NombreOccurrences)
        tup_elt = (elt,nbrOcc)
        ## pour éviter d'avoir des tuples en double
        ## vérifier si le tuple n'existe pas dans le résultat final
        if tup_elt not in res_tuples:
            ## s'il n'existe pas, le rajouter à la liste
            res_tuples.append(tup_elt)
    return res_tuples

## -- Appel de Fonction -- ##
# nombreOccurrence([-4,8,-3,2,1,2,7,9,-3,8,1])
nombreOccurrence(["a",3,4,"b","a",3])
```

```python
## Exercice 59

def unionListe(L1,L2,L3):
    ## Convertir toutes listes en ensembles
    ens1 = set(L1)
    ens2 = set(L2)
    ens3 = set(L3)
    
    ## Faire l'union des 3 listes
    unionEnsemble = ens1 | ens2 | ens3
    ## Convertir le résultat de l'union en liste
    L_union = list(unionEnsemble)
    ## Trier la liste dans un ordre croissant
    L_union.sort()
    
    return L_union

## --- Appel de Fonction --- ##
# unionListe([3,6,9,3],[1,0,3],[12,6,0])
unionListe([7,44,-3],[],[7,2,7])
```

```python
## Exercice 60

def calculPGCD(a,b):
    ## vérifier si les deux paramètres
    ## a et b sont positifs
    ## si ce n'est pas le cas, retourner une erreur
    assert(a>0 and b > 0)
    ## tant que b est différente de 0
    while b != 0:
        ## double affectation
        ## "a" on lui affecte "b"
        ## "b" on lui affecte le "reste de a divisé par b"
        a,b = b,a%b
    ## on retourne "a" à la fin
    return a

## --- Appel de Fonction --- ##
# calculPGCD(3,5)
calculPGCD(5,15)
```

```python
## Exercice 69

from datetime import datetime

def nbrJourHeure(dateDebut,dateFin):
    ## le format de la date
    format_date = "%Y/%m/%d"
    ## convertir la date de début de string à un format date
    date_debut_formate = datetime.strptime(dateDebut,format_date)
    ## convertir la date de fin de string à un format date
    date_fin_formate = datetime.strptime(dateFin,format_date)
    ## nombre de jour entre date début et date de fin
    nbrJour = (date_fin_formate - date_debut_formate).days
    ## calcul du nombre d'heures
    nbrHeure = nbrJour * 24
    
    return nbrJour,nbrHeure

## ---- Appel de Fonction ----- ##
# nbrJourHeure("2022/05/15","2022/06/20")
nbrJourHeure("2022/04/1","2022/04/27")
```

```python
## Exercice 70

## Importation des modules
import string
import random
## création d'une liste de caractères à partir de laquelle
## nous allons générer notre mot de passe aléatoirement
liste_caracteres = list(string.ascii_letters + string.digits + "!@#$%^&*()")

def genererMDP(caracteres, tailleMDP):
    ## mélanger aléatoirement les caractères
    random.shuffle(caracteres)
    ## création de la liste qui va contenir
    ## notre mot de passe
    mot_de_passe = []
    ## choisir aléatoirement les caractères du mot de passe
    ## jusqu'à atteindre la taille passée en paramètres
    for i in range(tailleMDP):
        carac_alea = random.choice(caracteres)
        ## ajouter le caractère choisi aléatoirement
        ## dans la liste mot_de_passe
        mot_de_passe.append(carac_alea)
    
    ## mélanger encore une fois notre liste
    ## qui va contenir notre mot de passe
    random.shuffle(mot_de_passe)
    ## convertir la liste en chaine de caractères
    ## sans espaces entre les caractères
    mot_de_passe = "".join(mot_de_passe)
    
    return mot_de_passe

## -- Appel de Fonction
genererMDP(liste_caracteres,10)  
```

```python
## Exercice 76

def join(L,caractere):
    ## initialiser une chaine de caractères vide
    chaine_carac = ""
    ## parcourir tous les index des éléments de L
    for i in range(len(L)):
        ## ajouter l'élément de L se trouvant à 
        ## l'index i à chaine_carac
        chaine_carac += L[i]
        ## si ce n'est pas le dernier élément
        ## on ajoute le caractère séparateur
        if i != len(L)-1:
            chaine_carac += caractere
    ## retourner la chaine obtenue à partir de L
    ## en utilisant le séparateur passé en 
    ## paramètres
    return chaine_carac

## --- Appel de Fonction -- ##
# join(["Bonjour","Aurélie"]," ")
join(["Salut","ça va?"], ",")
```

```python
## Exercice 77

def remplacer(phrase,ancienMot,nouveauMot):
    ## si le mot qu'on veut remplacer existe 
    ## dans la phrase passée en paramètres
    if ancienMot in phrase:
        ## extraire l'index de début de l'ancien mot
        debut_ancienMot_idx = phrase.index(ancienMot)
        ## extraire l'index de fin de l'ancien mot
        fin_ancienMot_idx = debut_ancienMot_idx + len(ancienMot)
        ## convertir la phrase en liste de caractères
        phrase_list = list(phrase)
        ## remplacer l'ancien mot par le nouveau
        phrase_list[debut_ancienMot_idx:fin_ancienMot_idx] = nouveauMot
        ## reconstitution de la nouvelle phrase en convertissant la liste 
        ## en chaine de caractères
        phrase = "".join(phrase_list)
    return phrase

## -- Appel de Fonction -- ##
# remplacer("Bonjour Aurélie","Aurélie","Mathilde")
remplacer("J'ai 50 ans", "50","35")
```

```python
## Exercice 78

def split(phrase,caractere):
    ## la variable qui va contenir notre liste
    phrase_list = []
    ## initialisation d'une variable temporaire
    mot_tmp = ""
    ## si le caractère se trouve dans la phrase
    ## passée en paramètres
    if caractere in phrase:
        ## parcourir tous les indices des éléments
        ## de la phrase
        for i in range(len(phrase)):
            ## tant que l'élément à l'index est
            ## différent du séparateur "caractere"
            if phrase[i] != caractere:
                ## alors on concaténe cet élément avec
                ## notre variable temporaire
                mot_tmp += phrase[i]
                ## si i est égal à l'index du dernier élément
                if i == len(phrase)-1:
                    ## ajouter le dernier mot
                    phrase_list += [mot_tmp]
            ## si l'élément à l'index i est égal au séparateur
            else:
                ## alors on ajoute le mot temporaire dans notre liste
                phrase_list += [mot_tmp]
                ## et on réinitialise le mot temporaire
                mot_tmp = ""
        return phrase_list
    # si le séparateur n'existe pas dans la phrase
    else:
        ## alors on ajoute la phrase entière dans la liste
        phrase_list += [phrase]
        return phrase_list

## --- Appel de Fonction -- ##
# split("Bonjour Aurélie", " ")
split("Salut, ça va", ",")
```

```python
## Exercice 88

def triCroissant(L):
    ## parcourir tous les index de la liste L
    for i in range(len(L)):
        ## pour les index supérieur à i
        for j in range(i+1, len(L)):
            ## si l'élément à l'index i est supérieur
            ## à l'élément se trouvant à j
            if L[i] > L[j]:
                ## échanger l'emplacement de ces deux nombres
                L[i], L[j] = L[j], L[i]
    return L

## --- Appel de Fonction -- ##
# triCroissant([6,1,9,-6,1,8,7])
triCroissant([-3,5.3,2,7,1,2.3,9.5])
```

```python
## Exercice 90

class Rectangle:
    ## constructeur pour initialiser automatiquement
    ## l'instance qu'on va créer
    def __init__(self,largeur,longueur):
        self.largeur = largeur
        self.longueur = longueur
    
    ## méthode qui permet de calculer le périmétre
    ## d'un rectangle
    def Perimetre(self):
        return 2*(self.largeur+self.longueur)
    ## méthode qui permet de calculer la surface
    ## d'un rectangle
    def Surface(self):
        return self.longueur * self.largeur
    
## --- Création d'une instance + utilisation de méthodes --- ##
## création d'une instance de la classe Rectangle
rectangle1 = Rectangle(10,15)
## afficher la largeur du rectangle1
print("La largeur du rectangle1 est : ",rectangle1.largeur)
## afficher la longueur du rectangle1
print("La longueur du rectangle1 est : ",rectangle1.longueur)
## calculer le périmètre du rectangle1
print("Le perimetre du rectangle1 est : ",rectangle1.Perimetre())
## calculer la surface du rectangle 1
print("La surface du rectangle1 est : ",rectangle1.Surface())
```

```python
## Exercice 91

## définition d'une classe qui hérite d'une autre classe
class Parallelepipede(Rectangle):
    ## constructeur
    def __init__(self,largeur,longueur,hauteur):
        ## initialisation des attributs de la classe Parent
        Rectangle.__init__(self,largeur,longueur)
        ## nouveau attribut de classe Parallelepipede
        self.hauteur = hauteur
    ## méthode qui calcule le volume
    def Volume(self):
        return self.longueur*self.largeur*self.hauteur
        
## --- Création d'une instance + utilisation de méthodes --- ##
## création d'une instance parallelepipede
parallelepipede1 = Parallelepipede(9,5,2)
## calcul du périmètre du parallelepipede
print("Perimètre du parallelepipede1 : ", parallelepipede1.Perimetre())
## calcul du volume
print("Volume du parallelepipede1 : ", parallelepipede1.Volume())
```

```python
## Exercice 94

class Point2D:
    ## Constructeur d'initialisation des attributs
    def __init__(self,x,y):
        self.x = x
        self.y = y
    ## surcharge de l'opérateur +
    def __add__(self,p):
        return self.x + p.x, self.y + p.y
    ## surcharge de l'opérateur -
    def __sub__(self,p):
        return self.x - p.x, self.y - p.y
    ## surcharge de l'opérateur x
    def __mul__(self,p):
        return self.x * p.x, self.y * p.y
    ## surcharge de l'opérateur /
    def __truediv__(self,p):
        return self.x/p.x, self.y/p.y

## -- Création d'instances et appel de méthodes -- ##
p1 = Point2D(3,2)
p2 = Point2D(1,4)
print("p1-p2 =",p1-p2)
print("p1+p2 =",p1+p2)
print("p1/p2 =",p1/p2)
```

```python
## Exercice 98

def ordonnerDict(d):
    ## liste qui va contenir les tuples
    liste_tuples = []
    ## liste des clés du dictionnaire
    cle_d = list(d.keys())
    ## liste des valeurs
    valeurs_d = list(d.values())
    ## boucler sur (cle,valeur)
    for cle,valeurs in zip(cle_d,valeurs_d):
        ## ajouter le tuple dans la liste
        liste_tuples.append((cle,valeurs))
    ## ordonner dans un ordre croissant suivant
    ## le premier élément du tuple
    liste_tuples.sort(key = lambda x : x[0])
    ## retourner un dictionnaire
    return dict(liste_tuples)

## -- Appel de Fonction -- ##
# ordonnerDict({"a":1,"d" : 7, "b" : 3, "c" : 2})
ordonnerDict({8:9,2:3,9:11})
```

```python
## Exercice 99

## somme d'une sous séquence liste
def SommeSeq(L,i,j):
    Lij = L[i:j+1]
    return sum(Lij)
        
def PlusGrandeSomme(L):
    ## initialisation var somme_max
    ## avec le premier élément de la liste L
    somme_max = L[0]
    ## parcourir tous les index de L
    for i in range(len(L)):
        ## pourcourir les index de L en partant de i
         for j in range(i, len(L)):
                ## appel de fonction pour sommer la séquence de i à j
                s = SommeSeq(L, i, j)
                ## si la somme trouvée est supérieur à celle que 
                ## nous avons initialisé
                if s > somme_max :
                    ## enregistrer la séquence trouvée dans une variable
                    seq = L[i:j+1]
                    ## changer le contenu de somme_max avec
                    ## le contenu de la nouvelle variable s
                    somme_max = s
    return somme_max,seq

## -- Appel de fonction -- ##                    
# PlusGrandeSomme([-8,-4,6,8,-6,10,-4,-4])
PlusGrandeSomme([-6,1,8,-7,1,9,-1,2])
```
